```python
#lets see something tricky codes 
#Following is the class with the constructor

class Car_: 
    
    fuel_type='Petrol'
    
    def __init__(self,name,brand,type):  #constructor (invoke when object created)
        self.name=name 
        self.brand=brand 
        self.type=type 
        
    def desc(self):
        print(f"Car Name: {self.name}") 
        print(f"Car Brand Name: {self.brand}") 
        print(f"Car Type: {self.type}") 
        print(f"Fuel Type: {self.fuel_type}") 
              
        

car_1=Car_("altroz","Tata","Hatchback")
car_1.desc()
```

## Inheritence

```python


#here we are going to inherit the class Car_ in Sportscar class 

class SportsCar(Car_):
    induction_system='Turbo Charged'
    
    def __init__(self,torque,horsepower,cooling_system,capacity):
        self.torque=torque 
        self.horsepower=horsepower 
        self.cooling_system= cooling_system 
        self.capacity=capacity 
        
    def description(self): 
        print(f"Car Torque: {self.torque}")
        print(f"Car Horsepower: {self.horsepower}")
        print(f"Car cooling_system: {self.cooling_system}")
        print(f"Car Capacity: {self.capacity}")
        print(f"Induction System: {self.induction_system}")
    
#lets create its object: 
sports_car1=SportsCar(200,250,'air_cooled',"1000cc") 

sports_car1.fuel_type #inherited class attribute  
#but here we can access the class attributes of parent class 
#because constructor of child class overrides the constructor of parent class 
```

## polymorphism


```python
# NOTE: Python doesn't supports method overloading like c++
# - It supports overriding, if parent and child have same name func with different params
# then the child's method overrides the parents methods 
#as you can see in the following example 

#Method overriding
class A: 
    name: str 
    
    def print_name(self): 
        print("name is ",self.name) 
        
    def demo(self):
        print("Class A") 
        
        
        
class B(A): 
    def print_name(self,count): 
        for i in range(count): 
            print("Name is ",self.name) 
    
    
    def demo(self):
        print('Class B')


obj1=B()
```

```python
obj1.name='aashish' 
obj1.print_name(3)
obj1.demo()
```

## Operator overloading

```python
#here we have to use special function(function that starts with __) 


class A: 
    def __init__(self,a,b): 
        self.a=a 
        self.b=b 
    
        
    def __add__(self,obj): 
        print("obj1 + obj2") 
        print("a = ", self.a+obj.a) 
        print("b = ",self.b+obj.b) 
    
    
    
    def __mul__(self,obj): 
        print("obj1 * obj2") 
        print("a = ",self.a*obj.a) 
        print('b = ',self.b*obj.b) 
        
    
    
    
    
obj1=A(213,6)
obj2=A(55,3)
```

```python
obj1+obj2
```

```python
obj1*obj2
```

## super()

- In Python OOP, the super() keyword/function is used to access methods and constructors of the parent class from the child class. It is mostly used with inheritance.

```python
class A: 
    def prin(self): 
        print('Hi my name is aashish') 
    
class B(A) : 
    def prin(self):
        super().prin() #invokes the parent class funtion
        print('I am a student') 
        
obj=B()
obj.prin()
```

```python
#constructor invocation 

class Person:
    def __init__(self, name):
        self.name = name
        print("Person constructor called")

class Student(Person):
    def __init__(self, name, roll):
        super().__init__(name)   # call parent constructor
        self.roll = roll
        print("Student constructor called")

s = Student("Aashish", 101)
```

```python
class A:
    def show(self):
        print("A")

class B(A):
    def show(self):
        super().show()
        print("B")

class C(A):
    def show(self):
        super().show()
        print("C")

class D(B, C):
    def show(self):
        super().show()
        print("D")
        
        
D.mro()
```
