```python
#Following is the class with the constructor

class Car: 
    
     
    
    def __init__(self,name,brand,type):  #constructor (invoke when object created)
        self.name=name 
        self.brand=brand 
        self.type=type 
        
    def desc(self):
        print(f"Car Name: {self.name}") 
        print(f"Car Brand Name: {self.brand}") 
        print(f"Car Type: {self.type}") 
        
    

#1 
car1=Car('Alto k10','maruti suzuki','hatchback') 

#2 
car2=Car(name="Tiago",type='hatchback',brand="Tata")

#lets invoke the desc() method 
car1.desc() #
car2.desc()

```

```python
#modification of the attributes of the object 
car1.brand='Suzuki'
```

```python
car1.name, car1.type,car1.brand #changed successfully
```

```python
#lets see something tricky codes 
#Following is the class with the constructor

class Car_: 
    
    fuel_type='Petrol'
    
    def __init__(self,name,brand,type):  #constructor (invoke when object created)
        self.name=name 
        self.brand=brand 
        self.type=type 
        
    def desc(self):
        print(f"Car Name: {self.name}") 
        print(f"Car Brand Name: {self.brand}") 
        print(f"Car Type: {self.type}") 
        print(f"Fuel Type: {self.fuel_type}") 
              
        

car_1=Car_("altroz","Tata","Hatchback")
car_1.desc()
```

```python
#accesing the class attributes 
Car_.fuel_type
```

```python
#lets modify this attribute 
Car_.fuel_type='Diesel'
```

```python
#will it reflect the values of the previously created object or not ? 
# ans = yes (because the attribute belongs to the class not the object)
car_1.fuel_type 
```

```python
#like if i modify the attribute of object1 then it will not reflect the class 
# eg
car_1.fuel_type='Petrol'
#it will not reflect the class attribute 
print(Car_.fuel_type)
print(car_1.fuel_type)
```

```python
#lets see inheritence 

#here we are going to inherit the class Car in Sports_car class 

class SportsCar(Car_):
    induction_system='Turbo Charged'
    
    def __init__(self,torque,horsepower,cooling_system,capacity):
        self.torque=torque 
        self.horsepower=horsepower 
        self.cooling_system= cooling_system 
        self.capacity=capacity 
        
    def description(self): 
        print(f"Car Torque: {self.torque}")
        print(f"Car Horsepower: {self.horsepower}")
        print(f"Car cooling_system: {self.cooling_system}")
        print(f"Car Capacity: {self.capacity}")
        print(f"Induction System: {self.induction_system}")
    
#lets create its object: 
sports_car1=SportsCar(200,250,'air_cooled',"1000cc") 

sports_car1.fuel_type #inherited class attribute  
#but here we can access the class attributes of parent class 
#because constructor of child class overrides the constructor of parent class 
```
