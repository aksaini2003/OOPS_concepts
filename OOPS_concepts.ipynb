```python
#Following is the class with the constructor

class Car: 
    
     
    
    def __init__(self,name,brand,type):  #constructor (invoke when object created)
        self.name=name 
        self.brand=brand 
        self.type=type 
        
    def desc(self):
        print(f"Car Name: {self.name}") 
        print(f"Car Brand Name: {self.brand}") 
        print(f"Car Type: {self.type}") 
        
    
#there are 2 ways for the creation of the object in the python 
#1 
car1=Car('Alto k10','maruti suzuki','hatchback') 

#2 
car2=Car(name="Tiago",type='hatchback',brand="Tata")

#lets invoke the desc() method 
car1.desc() #
car2.desc()

```

```python
#modification of the attributes of the object 
car1.brand='Suzuki'
```
